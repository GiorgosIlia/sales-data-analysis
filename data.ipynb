```python
import pandas as pd
import glob

import os 
```

```python
path = ("\\Users\\jegen\Desktop\\visualStudio\\data_science\\Sales_Data")
files = os.path.join(path, "Sales*.csv")
```

```python
files = glob.glob(files)
df = pd.concat(map(pd.read_csv, files), ignore_index=True) 
print(df)
df.to_csv("all_datas.csv")
```

```python
df.dropna(inplace=True)
total_prices = df.copy()
```

```python
total_prices1=total_prices.copy()
total_prices1=df.loc[:,["Order Date","Price Each","Quantity Ordered","Purchase Address","Product"]]
total_prices1
```

```python
total_prices2 = total_prices1.copy()
total_prices2["Month"] = total_prices1["Order Date"].str[0:2]
total_prices2
```

```python
total_prices3 = total_prices2.copy()
total_prices3 = total_prices3[total_prices3["Order Date"].str[0:2] != "Or"]
total_prices3
```

```python
total_prices4 = total_prices3.copy()
total_prices4["Month"] = total_prices4["Month"].astype("int32")
total_prices4["Price Each"] = total_prices4["Price Each"].astype("float64")
total_prices4["Quantity Ordered"]= total_prices4["Quantity Ordered"].astype("int32")
total_prices4["Total Cost"]= total_prices4["Price Each"]* total_prices4["Quantity Ordered"]
total_prices4
```

```python
results = total_prices4.groupby("Month").sum()
results
```

```python
results = total_prices4.groupby("Month").sum()
import matplotlib.pyplot as plt
months = range(1,13)
names = plt.bar(months,results["Total Cost"])
plt.ylabel("Price in USD ($)")
plt.xlabel("Months")
plt.show()
```

```python
def get_city(address):
    return address.split(",")[1]
def get_state(address):
    return address.split(",")[2].split(" ")[1]

total_prices4["City"] = total_prices4["Purchase Address"].apply(lambda x: get_city(x) + " ("+ get_state(x)+ ")")
total_prices4
```

```python
final = total_prices4.groupby("City").sum()
final
```

```python
final.drop(columns=['Month','Quantity Ordered'])
```

```python
cities = [city for city, final in final.groupby("City")]

names = plt.bar(cities,final["Total Cost"])
plt.xticks(cities,rotation="vertical")
plt.ylabel("Sales")
plt.xlabel("Cities")
plt.show()
```

```python
def get_hours(hours):
    yolo = hours.split(":")[0]
    return yolo.split(" ")[1]

total_prices4["Time"] = total_prices4["Order Date"].apply(lambda x:  get_hours(x))
total_prices4
```

```python
hour = [time for time, df in total_prices4.groupby("Time")]

plt.plot(hour,total_prices4.groupby(["Time"]).count())
plt.grid()
plt.ylabel("Sales")
plt.xlabel("Time")
plt.show()
```

```python
sold_together["Grouped"] = sold_together.groupby("Order ID")["Product"].transform(lambda x: "," .join(x))

sold_together.head()
```

```python
sold_together = sold_together[["Order ID","Grouped"]].drop_duplicates()


sold_together.head(20)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in sold_together["Grouped"]:
    rowlist = row.split(",")
    count.update(Counter(combinations(rowlist,2)))


count.most_common(20)
```
